```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("level_salary_dataset.csv")
```

```python
dataset.head(3)
```

```python
dataset.corr()
```

```python
dataset.isnull().sum()
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.xlabel("Laevl")
plt.ylabel("Salary")
plt.show()
```

```python
x = dataset[["Level"]]
y = dataset[["Salary"]]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=2)
pf.fit(x)
x = pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#y = m1x1+m2x2^2+c
#y = -85861.40536579*x1 + 4296.25960178+x2^2 - 412320.92364208
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
prd = lr.predict(x)
```

```python
plt.scatter(dataset["Level"],dataset["Salary"])
plt.plot(dataset["Level"],prd,c='red')
plt.xlabel("Laevl")
plt.ylabel("Salary")
plt.legend(["org","prd"])
plt.show()
```
